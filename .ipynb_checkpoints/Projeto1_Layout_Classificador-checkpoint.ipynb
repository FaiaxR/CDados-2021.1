### Montando um Classificador Naive-Bayes

Considerando apenas as mensagens da planilha Treinamento, ensine  seu classificador.

```python
#Fazendo uma limpeza inicial dos dados

#Removendo pontua√ß√µes
def removepont(text):
    punctuation = '[!-.:?;_,<>/\][)(#%$|\n]' 
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern,'', text)
    return text_subbed

#Removendo men√ß√µes
def removemencao(text):
    pattern = re.compile("@[A-Za-z0-9]+")
    return re.sub(pattern,'',text)

#Separando os emojis
def separaemoji(text):
    text_emoji_split = emoji.get_emoji_regexp().split(text)
    separa1 = [substr.split() for substr in text_emoji_split]
    separa2 = functools.reduce(operator.concat, separa1)
    separa3 = ' '.join(separa2)
    return separa3
    
train.Treinamento = train.Treinamento.apply(removepont)
train.Treinamento = train.Treinamento.apply(removemencao)
train.Treinamento = train.Treinamento.apply(separaemoji)
#Mais pra frente ser√° feita mais uma rodada de limpeza
```

```python
train.head(10)
#Como se pode ver abaixo, os emojis ficaram separados na linha 9
```

Nota: N√£o √© necess√°rio fazer uma limpeza das bases mais aprofundadas - os tweets j√° s√£o baixados sem letras mai√∫sculas

```python
#Alterando a tabela para trabalhar com categorias
train['B1'] = train['B1'].astype('category')
train['B1'].cat.categories = ['Irrelevante', 'Relevante']
train['B1'].cat.categories
```

```python
#Separando os tweets relevantes e irrelevantes em diferentes dataframes
rel = train.loc[train.B1=='Relevante',:]
irrel = train.loc[train.B1=='Irrelevante',:]
```

```python
rel
```

```python
#Criando uma lista com todas as palavras de cada categoria
palavras_rel = []
palavras_irrel = []

for tweet in (rel.Treinamento.to_list()):
    for palavra in tweet.split():
        palavras_rel.append(palavra)
        
for tweet in (irrel.Treinamento.to_list()):
    for palavra in tweet.split():
        palavras_irrel.append(palavra)
```

```python
#Criando o dicion√°rio com todas as palavras do nosso classificador
dicionario = palavras_rel + palavras_irrel
#Transformando dicionario em uma Series pra podermos aplicar value_counts()
dicionario_serie = pd.Series(dicionario)
#Procurando stopwords (palavras frequentes sem significado pro classificador)
dicionario_serie.value_counts().head(50)
```

```python
#Criando uma lista pras stop words mais frequentes no dicion√°rio
stopwords = ['o','e','que','de','a','no','do','um','eu','pra','para','em','com','tem','mas','na','uma','se','mais','meu','da','me','s√≥','os','por','q','ou','tenho','isso','quando','ele','j√°','ao','as']
#Fazendo a segunda rodada de limpeza e removendo stopwords
for palavra in palavras_rel:
    if palavra in stopwords:
        palavras_rel.remove(palavra)
for palavra in palavras_irrel:
    if palavra in stopwords:
        palavras_irrel.remove(palavra)
```

```python
#Transformando listas de palavras em Series para aplicar value_counts
serie_rel = pd.Series(palavras_rel)
serie_irrel = pd.Series(palavras_irrel)
#Obtendo tabelas de frequ√™ncia absoluta e relativa
tabela_rel = serie_rel.value_counts()
tabela_rel_relativa = serie_rel.value_counts(True)
tabela_irrel = serie_irrel.value_counts()
tabela_irrel_relativa = serie_irrel.value_counts(True)
```

```python
#Recriando o dicion√°rio agora sem stopwords
dicionario = palavras_rel + palavras_irrel
#Fazendo o mesmo processo de cria√ß√£o de tabelas
serie_dicionario = pd.Series(dicionario)
tabela_dicionario = serie_dicionario.value_counts()
tabela_dicionario_relativa = serie_dicionario.value_counts(True)
```

```python
#Clarificando os elementos que temos at√© agora e que ser√£o utilizados no classificador:
P_rel = tabela_rel.sum()/tabela_dicionario.sum()
P_irrel = tabela_irrel.sum()/tabela_dicionario.sum()
```

```python
#Podemos usar uma frase gen√©rica para fazer o teste
tweet = "@fulano o xbox √© muito bom! acho que o controle dele √© muito melhor do que o do playstation"
```

```python
#Fazendo a fun√ß√£o comparadora.
def comparar(texto):
    #Limpando o tweet e separando as palavras individualmente
    texto_limpo = removemencao(texto)
    texto_limpo = removepont(texto_limpo)
    texto_limpo = separaemoji(texto_limpo)
    lista_texto = texto_limpo.split()
            
    #Procurando as probabilidades dado que o texto √© relevante/irrelevante e probabilidade do texto dado nosso dicion√°rio
    #Nessa etapa ocorre a Suaviza√ß√£o de Laplace para evitar probabilidades = 0
    P_texto_dado_rel = 1
    P_texto_dado_irrel = 1
    for palavra in lista_texto:
        if palavra not in tabela_rel:
            P_texto_dado_rel = P_texto_dado_rel/(tabela_rel.sum() + len(lista_texto))
        else:
            P_texto_dado_rel = P_texto_dado_rel*(tabela_rel[palavra] + 1)/(tabela_rel.sum() + len(lista_texto))
        if palavra not in tabela_irrel:
            P_texto_dado_irrel = P_texto_dado_irrel/(tabela_irrel.sum() + len(lista_texto))
        else:
            P_texto_dado_irrel = P_texto_dado_irrel*(tabela_irrel[palavra] + 1)/(tabela_irrel.sum() + len(lista_texto))
    
    #Calculando as probabilidades de ser relevante/irrelevante dado a frase
    P_rel_dado_texto = P_texto_dado_rel * P_rel
    P_irrel_dado_texto = P_texto_dado_irrel * P_irrel
    #Finalmente, retornando se o tweet √© relevante ou irrelevante
    if P_rel_dado_texto > P_irrel_dado_texto:
        return "Relevante"
    if P_irrel_dado_texto > P_rel_dado_texto:
        return "Irrelevante"
```

### Concluindo

<em><b>Discutindo a performance do comparador</b></em> <br>
Como pode-se perceber, o classificador criado conseguiu classificar corretamente apenas 60% dos tweets, o que a primeira vista parece pouco satisfat√≥rio. 
Contudo, se os tweets relevantes fossem repassados para an√°lise, levando em considera√ß√£o os falsos positivos, o analista teria que lidar apenas com aproximadamente 20% de tweets irrelevantes, o que parece bem mais adequado do que lidar com uma porcentagem de aproximadamente 50% de tweets irrelevantes da base de dados bruta. √â verdade que h√° perda de informa√ß√£o (alguns tweets relevantes s√£o erronimanete classificados - os falsos negativos) mas o comparador serve a sua fun√ß√£o de diminuir a quantidade de tweets irrelevantes de qualquer maneira.

<em><b>Limita√ß√µes do comparador</b></em><br>
Por ser um comparador ing√™nuo, as palavras de um tweet s√£o interpretadas como idependentes uma das outras. Isso significa que duplos negativos (como "Nunca encontrei nada"), que gramaticalmente n√£o faz sentido, recupera o seu valor por serem entendidos por seus componentes individuais: nunca e nada refor√ßam a ideia de nega√ß√£o nesse caso, ao inv√©s de se cancelarem.<br>
Em rela√ß√£o a ironias, como deseja-se obter tweets que discutam a imagem do produto xbox, n√£o importa para n√≥s se o autor do texto est√° sendo ir√¥nico ou n√£o. Se quis√©ssemos avaliar apenas os tweets que falassem bem do console (para avaliar o que a comunidade entende como positivo, por exemplo), por outro lado, n√£o temos o mesmo benef√≠cio. A programa√ß√£o vai sempre entender o texto literal, j√° que n√£o h√° um indicador de ironia que acompanhe os tweets. Frase do tipo "O meu xbox acabou de quebrar sozinho. Eu amo essa empresa" apresentaria um problema para o classificador: nesse caso um tweet que apresenta algo negativo sobre o console poderia aparecer como positivo (j√° que "amo" est√° tipicamente relacionado √† algo positivo). Felizmente, esse n√£o √© o caso, ent√£o ironias n√£o seriam prejudiciais para o nosso classificador

<em><b>Poss√≠veis Melhorias</b></em><br>
Caso se desejasse melhorar o algor√≠timo, poder√≠amos fazer tanto uma mudan√ßa na base de dados de treinamento quanto na programa√ß√£o do classificador.<br>
Quanto √† base de dados de treino, uma possibilidade √© aumentar a quantidade de tweets classificados "na coragem", dando ao computador um repert√≥rio maior do que pode ser considerado relevante ou irrelevante. Dessa forma, mais palavras seriam adicionadas ao nosso dicion√°rio, deixando nosso comparador mais inteligente.<br>
Quanto √† programa√ß√£o do classificador, temos duas poss√≠veis mudan√ßas. Uma op√ß√£o √© fazer um processo de lemmatiza√ß√£o (lemmatization) das palavras. Este processo busca agrupar palavras de mesmo radical e que, portanto, tem um significado parecido. Dessa forma, palavras como "jogos", "jogar" e "jogamos" passam por uma transforma√ß√£o e viram todas o que elas tem em comum ("jog"), afim de aumentar a ocorr√™ncia dessa palavra no dicion√°rio e facilitando a classifica√ß√£o (mais informa√ß√µes a respeito desse processo est√° descrito no estudo [Naive Bayes and Text Classification](https://arxiv.org/pdf/1410.5329.pdf)). A outra op√ß√£o √© usar "n-grams", que juntam palavras que tem um significado importante quando juntas, diminuindo a "ingenuidade" do nosso classificador.

<em><b>Outros cen√°rios para o uso de um classificador Naive-Bayes</b></em><br>
Esse tipo de classificador √© tipicamente utilizado em filtros de spam de e-mail, criando um vocabul√°rio de palavras tipicamente utilizadas em e-mails denunciados como spam e usando o algor√≠timo para separar e-mails com essa palavra.<br>
Uma outra aplica√ß√£o √© em √©poca de elei√ß√£o. Pode ser interessante verificar a diferen√ßa no engajamento web de diferentes candidatos. Na elei√ß√£o de 2018, por exemplo, poderia-se medir o qu√£o maior √© o engajamento online de eleitores do candidato Jair Bolsonaro contra candidatos mais tradicionalistas, como o Geraldo Alckmin, filtrando tweets de eleitores e diferenciando-os de simples coment√°rios de pol√≠tica.

```python
#ITEM 3 - Montando o classificador baseado em uma nova tabela de treinamento

#Selecionando 300 tweets aleat√≥riamente e salvando em um DataFrame de treino
train = df.sample(300)
#Limpeza inicial
train.Tweet = train.Tweet.apply(removepont)
train.Tweet = train.Tweet.apply(removemencao)
train.Tweet = train.Tweet.apply(separaemoji)
#Separando os tweets relevantes e irrelevantes em diferentes dataframes
rel = train.loc[train.B1=='Relevante',:]
irrel = train.loc[train.B1=='Irrelevante',:]
#Criando uma lista com todas as palavras de cada categoria
palavras_rel = []
palavras_irrel = []

for tweet in (rel.Tweet.to_list()):
    for palavra in tweet.split():
        palavras_rel.append(palavra)
        
for tweet in (irrel.Tweet.to_list()):
    for palavra in tweet.split():
        palavras_irrel.append(palavra)
        
#Fazendo a segunda rodada de limpeza e removendo stopwords
for palavra in palavras_rel:
    if palavra in stopwords:
        palavras_rel.remove(palavra)
for palavra in palavras_irrel:
    if palavra in stopwords:
        palavras_irrel.remove(palavra)
        
#Transformando listas de palavras em Series para aplicar value_counts
serie_rel = pd.Series(palavras_rel)
serie_irrel = pd.Series(palavras_irrel)
#Obtendo tabelas de frequ√™ncia absoluta e relativa
tabela_rel = serie_rel.value_counts()
tabela_rel_relativa = serie_rel.value_counts(True)
tabela_irrel = serie_irrel.value_counts()
tabela_irrel_relativa = serie_irrel.value_counts(True)
#Criando o dicion√°rio
dicionario = palavras_rel + palavras_irrel
#Fazendo o mesmo processo de cria√ß√£o de tabelas
serie_dicionario = pd.Series(dicionario)
tabela_dicionario = serie_dicionario.value_counts()
tabela_dicionario_relativa = serie_dicionario.value_counts(True)
#Clarificando os elementos que temos at√© agora e que ser√£o utilizados no classificador:
P_rel = tabela_rel.sum()/tabela_dicionario.sum()
P_irrel = tabela_irrel.sum()/tabela_dicionario.sum()

#ITEM 4 - Aplicando o classificador √† planilha de teste

#Recolhendo a planilha de teste
test = df.drop(train.index)
#Pegando os tweets sem classifica√ß√£o
tweets = test.loc[:,'Tweet']
#Fazendo a classifica√ß√£o dos tweets
tweets_classificados = []
for linha in tweets:
    tweets_classificados.append(comparar(linha))
#Colecionando as classifica√ß√µes reais
classe = test.loc[:,'B1']
classe_real = []
for linha in classe:
    classe_real.append(linha)
#Adicionando √† tabela
result = pd.DataFrame(tweets_classificados)
result.columns = ['Bayes']
result['Real'] = pd.Series(classe_real)
#Verificando taxa de verdadeiros positivos
tab_Vpos = result.loc[(result.Real=='Relevante') & (result.Bayes=='Relevante') ]
Vpos = 100*len(tab_Vpos)/len(result)
#Verificando taxa de verdadeiros negativos
tab_Vneg = result.loc[(result.Real=='Irrelevante') & (result.Bayes=='Irrelevante') ]
Vneg = 100*len(tab_Vneg)/len(result)
linha_resultado = [Vpos,Vneg]
linha_resultado
```

```python
k = 0
matriz = []
while k <100:
    #ITEM 3 - Montando o classificador baseado em uma nova tabela de treinamento

    #Selecionando 300 tweets aleat√≥riamente e salvando em um DataFrame de treino
    train = df.sample(300)
    #Limpeza inicial
    train.Tweet = train.Tweet.apply(removepont)
    train.Tweet = train.Tweet.apply(removemencao)
    train.Tweet = train.Tweet.apply(separaemoji)
    #Separando os tweets relevantes e irrelevantes em diferentes dataframes
    rel = train.loc[train.B1=='Relevante',:]
    irrel = train.loc[train.B1=='Irrelevante',:]
    #Criando uma lista com todas as palavras de cada categoria
    palavras_rel = []
    palavras_irrel = []

    for tweet in (rel.Tweet.to_list()):
        for palavra in tweet.split():
            palavras_rel.append(palavra)

    for tweet in (irrel.Tweet.to_list()):
        for palavra in tweet.split():
            palavras_irrel.append(palavra)

    #Fazendo a segunda rodada de limpeza e removendo stopwords
    for palavra in palavras_rel:
        if palavra in stopwords:
            palavras_rel.remove(palavra)
    for palavra in palavras_irrel:
        if palavra in stopwords:
            palavras_irrel.remove(palavra)

    #Transformando listas de palavras em Series para aplicar value_counts
    serie_rel = pd.Series(palavras_rel)
    serie_irrel = pd.Series(palavras_irrel)
    #Obtendo tabelas de frequ√™ncia absoluta e relativa
    tabela_rel = serie_rel.value_counts()
    tabela_rel_relativa = serie_rel.value_counts(True)
    tabela_irrel = serie_irrel.value_counts()
    tabela_irrel_relativa = serie_irrel.value_counts(True)
    #Criando o dicion√°rio
    dicionario = palavras_rel + palavras_irrel
    #Fazendo o mesmo processo de cria√ß√£o de tabelas
    serie_dicionario = pd.Series(dicionario)
    tabela_dicionario = serie_dicionario.value_counts()
    tabela_dicionario_relativa = serie_dicionario.value_counts(True)
    #Clarificando os elementos que temos at√© agora e que ser√£o utilizados no classificador:
    P_rel = tabela_rel.sum()/tabela_dicionario.sum()
    P_irrel = tabela_irrel.sum()/tabela_dicionario.sum()

    #ITEM 4 - Aplicando o classificador √† planilha de treino

    #Recolhendo a planilha de treino
    test = df.drop(train.index)
    #Pegando os tweets sem classifica√ß√£o
    tweets = test.loc[:,'Tweet']
    #Fazendo a classifica√ß√£o dos tweets
    tweets_classificados = []
    for linha in tweets:
        tweets_classificados.append(comparar(linha))
    #Colecionando as classifica√ß√µes reais
    classe = test.loc[:,'B1']
    classe_real = []
    for linha in classe:
        classe_real.append(linha)
    #Adicionando √† tabela
    result = pd.DataFrame(tweets_classificados)
    result.columns = ['Bayes']
    result['Real'] = pd.Series(classe_real)
    #Verificando taxa de verdadeiros positivos
    tab_Vpos = result.loc[(result.Real=='Relevante') & (result.Bayes=='Relevante') ]
    Vpos = 100*len(tab_Vpos)/len(result)
    #Verificando taxa de verdadeiros negativos
    tab_Vneg = result.loc[(result.Real=='Irrelevante') & (result.Bayes=='Irrelevante') ]
    Vneg = 100*len(tab_Vneg)/len(result)
    linha_resultado = [Vpos,Vneg]
    matriz.append(linha_resultado)
    k = k + 1
```
